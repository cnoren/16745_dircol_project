# Project Code: Hybrid Trajectory Optimization for Autonomous Vehicles with Velocity-Loss Collision Constraints


In this problem, I am using a direct method to optimize a driving trajectory for a simple bicycle model using a hybrid dynamics/multi-phase formulation. The "hybrid" part of this problem comes in from specifying an "event" in the hybrid dynamics that corresponds to a collision. We will solve this problem using IPOPT.


## The Dynamics
I will model the dynamics of the car as with a simple nonlinear bicycle model, with the following state and control:

$$\begin{align} x = \begin{bmatrix} p_x \\ p_y \\ \theta \\ \delta \\ v \end{bmatrix}, &\quad \quad u = \begin{bmatrix} a \\ \dot{\delta} \end{bmatrix}  \end{align} $$ 

where $p_x$ and $p_y$ describe the 2d position of the bike, $\theta$ is the orientation, $\delta$ is the steering angle, and $v$ is the velocity. The controls for the bike are acceleration $a$, and steering angle rate $\dot{\delta}$.  

```julia
function car_dynamics(x::Vector, u::Vector)::Vector
    # Nonlinear Bicycle Model's Continuous Time Dynamics 
    px, py, θ, δ, v = x 
    a, δdot, h = u 
    
    β = atan(1.485 * δ, 2.97)
    s,c = sincos(θ + β)
    ω = v*cos(β)*tan(δ) / 2.97
    
    vx = v*c 
    vy = v*s 
    
    xdot = [
        vx,
        vy,
        ω,
        δdot, 
        a
    ]
    
    return xdot 
end

function collision_velocity_loss(r)
    return r
end

function collision_jump_map(x)
    # The Vehicle experiences a collision
    xn = x
    xn[5] = x[5] - 1
    return xn
end
```

```julia
# DIFFERENT INTEGRATION STRATEGIES


function rk4(ode::Function, x::Vector, u::Vector)::Vector
    # USING RK4 WITH "FREE TIME" FORMULATION, HERE THE FREE TIME PARAMETER CORRESPONDS TO u[3].
    k1 = u[3] * ode(x,        u)
    k2 = u[3] * ode(x + k1/2, u)
    k3 = u[3] * ode(x + k2/2, u)
    k4 = u[3] * ode(x + k3,   u)
    return x + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
end

function hermite_simpson(ode::Function, x1::Vector, x2::Vector, u)::Vector
    xk12 = 0.5 * (x1 + x2) + 0.125 * u[3] *(ode(x1, u) - ode(x2, u))
    ẋm = (-3/(2.0*u[3]))*(x1 - x2) - 0.25*(ode(x1, u) + ode(x2, u))
    fm = ode(xk12, u)
    return fm -  ẋm
end
```

Time to setup and solve a constrained nonlinear program.  The optimization problem looks complicated but each piece should make sense and be relatively straightforward to implement. First we have the following LQR cost function that will track $x_{ref}$ (`Xref`) and $u_{ref}$ (`Uref`):

$$  J(x_{1:N},u_{1:N-1}) = \sum_{i=1}^{N-1} \bigg[ \frac{1}{2} (x_i - x_{ref,i})^TQ(x_i - x_{ref,i}) + \frac{1}{2} (u_i-u_{ref,i})^TR(u_i-u_{ref,i}) \bigg] + \frac{1}{2}(x_N - x_{ref,N})^TQ_f(x_N - x_{ref,N})$$

Which goes into the following full optimization problem:
   $$ \begin{align} \min_{x_{1:N},u_{1:N-1}} \quad & J(x_{1:N},u_{1:N-1}) & \\ 
 \text{st} \quad &  x_1 = x_{ic} & \tag{1}\\
 \quad & x_N = x_{g} &\tag{2}\\
 &  x_{k+1} = f(x_k,u_k)  &\text{for } k \not\in  \text{collision} \\
  &  x_{k+1} = g(f(x_k,u_k))  &\text{for } k \in   \text{collision} \\
  & x_l \geq -x \\
  & x_u \geq x\\
  & u_l \geq -u\\
  & u_u \geq u  
 \end{align}$$
 
 Each constraint is now described, with the type of constraint for `fmincon` in parantheses:
 1. Initial condition constraint **(equality constraint)**. 
 2. Terminal condition constraint **(equality constraint)**. 
 3. Driving discrete dynamics **(equality constraint)**. .
 4. Control and State **(primal bounds)**.

And here we have the list of mathematical functions to the Julia function names:
- $f$ is `car_dynamics` + `Hermite Simpson`
- $g$ is `collision_jump_map`

```julia
function reference_trajectory(nx, xic, xg, dt, N)
    # creates a reference Xref and Uref for walker 
    
    Uref = [[0.01; 0.01; 0.1] for i = 1:(N-1)]
    
    Xref = [zeros(nx) for i = 1:N]
    
    horiz_v = (3/N)/dt 
    xs1 = range(xic[1], 10, length = 50)
    xs2 = range(10, xg[1], length = 50)
    ys = range(xic[2], xg[2], length = N)
    ts = range(xic[3], xg[3], length = N)
    vs1 = range(xic[5], 6, length = 50)
    vs2 = range(6, xg[5], length = 50)
    Xref[1] = 1*xic 
    Xref[N] = 1*xg 
        #x     px py  y  d  v
    for i = 2:(N-1) 
        if (i<50)
            Xref[i] = [xs1[i],ys[i], ts[i], 0, vs1[i]]
        elseif (i==50)
            Xref[i] = [10; 0; 0; 0; 6]
        else
            Xref[i] = [xs2[i-50],ys[i-50], ts[i-50], 0, vs2[i-50]]
        end
    end
        
    return Xref, Uref
end
```

To solve this problem with Ipopt and `fmincon`, we are going to concatenate all of our $x$'s and $u$'s into one vector (same as HW3Q1):

$$ Z = \begin{bmatrix}x_1 \\ u_1 \\ x_2 \\ u_2 \\ \vdots \\ x_{N-1} \\ u_{N-1} \\ x_N \end{bmatrix} \in \mathbb{R}^{N \cdot nx + (N-1)\cdot nu} $$ 

where $x \in \mathbb{R}^{nx}$ and $u \in \mathbb{R}^{nu}$. Below we will provide useful indexing guide in `create_idx` to help you deal with $Z$.  Remember that the API for `fmincon` (that we used in HW3Q1) is the following:
$$ \begin{align} \min_{z} \quad & \ell(z) & \text{cost function}\\ 
 \text{st} \quad & c_{eq}(z) = 0 & \text{equality constraint}\\
 & c_L \leq c_{ineq}(z) \leq c_U & \text{inequality constraint}\\
 & z_L \leq z \leq z_U & \text{primal bound constraint}
 \end{align}$$
 
 Template code has been given to solve this problem but you should feel free to do whatever is easiest for you, as long as you get the trajectory shown in the animation `walker.gif` and pass tests. 

```julia
# feel free to solve this problem however you like, below is a template for a 
# good way to start. 

function create_idx(nx,nu,N)
    # create idx for indexing convenience
    # x_i = Z[idx.x[i]]
    # u_i = Z[idx.u[i]]
    # and stacked dynamics constraints of size nx are 
    # c[idx.c[i]] = <dynamics constraint at time step i>
    #
    # feel free to use/not use this 
    
    # our Z vector is [x0, u0, x1, u1, …, xN]
    nz = (N-1) * nu + N * nx # length of Z 
    x = [(i - 1) * (nx + nu) .+ (1 : nx) for i = 1:N]
    u = [(i - 1) * (nx + nu) .+ ((nx + 1):(nx + nu)) for i = 1:(N - 1)]
    
    # constraint indexing for the (N-1) dynamics constraints when stacked up
    c = [(i - 1) * (nx) .+ (1 : nx) for i = 1:(N - 1)]
    nc = (N - 1) * nx # (N-1)*nx 
    
    return (nx=nx,nu=nu,N=N,nz=nz,nc=nc,x= x,u = u,c = c)
end

function car_cost(params::NamedTuple, Z::Vector)::Real
    # cost function 
    idx, N, xg = params.idx, params.N, params.xg
    Q, R, Qf = params.Q, params.R, params.Qf
    Xref,Uref = params.Xref, params.Uref
    
    # CAN DONE: input vehicle tracking LQR cost 
    J = 0 
    for i = 1:(N-1)
        xi = Z[idx.x[i]]
        ui = Z[idx.u[i]]
       
        J += (0.5*transpose(xi-Xref[i]) * Q * (xi-Xref[i]) + 0.5*transpose(ui-Uref[i]) * R * (ui-Uref[i]))

    end
    
    # taken from earlier: dont forget terminal cost - Thanks! 
    xN = Z[idx.x[N]]
    J += 0.5*transpose(xN-Xref[N]) * Qf * (xN-Xref[N])
    
    return J 
end

function car_dynamics_constraints(params::NamedTuple, Z::Vector)::Vector
    idx, N, dt = params.idx, params.N, params.dt
        
    # create c in a ForwardDiff friendly way (check HW0)
    c = zeros(eltype(Z), idx.nc)
    
    collision_occured = false
    
    for i = 1:(N-1)
        xi = Z[idx.x[i]]
        ui = Z[idx.u[i]] 
        xip1 = Z[idx.x[i+1]]
        
        # USING RK4
        #c[idx.c[i]] = rk4(car_dynamics, xi, ui, dt) - xip1
        # USING HERMITE SIMPSON
                
        #d = sqrt((xi[1]-10)^2 + xi[2]^2)
        
        #if (d < 0.1) && !(collision_occured)
                #c[idx.c[i]] = hermite_simpson(car_dynamics, xi, collision_jump_map(xip1), ui)
                #c[idx.c[i]] = collision_jump_map(rk4(car_dynamics, xi, ui)) - xip1
                #collision_occured = true
                #c[idx.c[i]] = hermite_simpson(car_dynamics, xi, xip1, ui)
        if (i == 50)
                c[idx.c[i]] = collision_jump_map(rk4(car_dynamics, xi, ui)) - xip1
        else
                c[idx.c[i]] = rk4(car_dynamics, xi, ui) - xip1
                #c[idx.c[i]] = hermite_simpson(car_dynamics, xi, xip1, ui)
        end
    end
    return c 
end

    
function car_equality_constraint(params::NamedTuple, Z::Vector)::Vector
    N, idx, xic, xg = params.N, params.idx, params.xic, params.xg 
    
    # CAN DONE: stack up all of our equality constraints 
    
    # should be length 2*nx + (N-1)*nx + N 
    # inital condition constraint (nx)       (constraint 1)
    # terminal constraint         (nx)       (constraint 2)
    # dynamics constraints        (N-1)*nx   (constraint 3-6)
    # stance constraint           N          (constraint 7-8)

    # CAN DONE: return all of the equality constraints 
    ind_x0 = first(params.idx.x) # INITIAL CONDITION
    ind_xf = last(params.idx.x)  # FINAL GOAL
    return [Z[ind_x0] - params.xic; Z[ind_xf] - params.xg; Z[idx.x[50]] - params.xcol; car_dynamics_constraints(params, Z)]
end

function car_inequality_constraint(params::NamedTuple, Z::Vector)::Vector
    idx, N, dt = params.idx, params.N, params.dt
        
    # create c in a ForwardDiff friendly way (check HW0)
    c = zeros(eltype(Z), 1)
    
    # TODO: add the length constraints shown in constraints (9-10)
    # there are 2*N constraints here 
    c[1] = params.xcol[5]
    return c
end
```

```julia
    # problem size 
    nx = 5 
    nu = 2 
    tf2 = 10
    dt = 0.1 # 0.05 is also reasonable... 
    t_vec2 = tf:dt:(tf+tf2)
    N2 = length(t_vec)
    
    # constants 
    rad2deg = 180/pi
    deg2rad = pi/180

    # constraints
    min_v = 0  # Right now, no going backwards
    max_v = 10 # Max Limiter on the Car --> 10 m/s
    max_steer = 35 *deg2rad # Vehicle can't steer any more than this
    min_steer = -1 * max_steer # Vehicle can't steer any more than this
    max_accel = 0.8  # Max Acceleration
    min_accel = -1 * max_accel
    max_strrate = 5 * deg2rad # Highest rate of turn for the wheel
    min_strrate = -1 * max_strrate # Highest rate of turn for the wheel

    # initial and goal states 
    #x     px py  y  d  v
xic2 = Xm[:,end] - [0; 0; 0; 0; 0.0] 
xg2 =  Xm[:,end] + [35; -10; 0; 0; -Xm[5,end]]
xr3 = [Xm[1,end]; Xm[2,end]; Xm[3,end]; Xm[4,end]; Xm[5,end]]
xr4 =  [Xm[1,end] + 35; Xm[2,end] - 10; 0; 0; 0]

    # index sets 
    M1 = vcat([ (i-1)*10      .+ (1:5)   for i = 1:5]...)
    M2 = vcat([((i-1)*10 + 5) .+ (1:5)   for i = 1:4]...)
    J1 = [5,15,25,35]
    J2 = [10,20,30,40]
    
    # reference trajectory 
    Xref2, Uref2 = reference_trajectory(nx, xr3, xr4, dt, N2)
    
    # LQR cost function (tracking Xref, Uref)
    Q = diagm([1.0; 1.0; 1.0; 0.0; 1.0]);
    R = diagm(fill(1e-3,2))
    Qf = 1*Q;
    
    # create indexing utilities 
    idx2 = create_idx(nx,nu,N)
    
    # put everything useful in params 
    params2 = (
        model = model, 
        nx = nx,
        nu = nu,
        tf = tf2, 
        dt = dt, 
        t_vec = t_vec2,
        N = N, 
        M1 = M1, 
        M2 = M2,
        J1 = J1, 
        J2 = J2,
        xic = xic2, 
        xg = xg2,
        idx = idx2,
        Q = Q, R = R, Qf = Qf,
        Xref = Xref2, 
        Uref = Uref2
    )
    
    # CAN DONE: primal bounds (constraint 11)
    x_l = -Inf*ones(idx.nz) # update this 
    x_u =  Inf*ones(idx.nz) # update this
        
    for i=1:N
        x_l[(nx+nu)*(i-1)+4] = min_steer
        x_u[(nx+nu)*(i-1)+4] = max_steer    
        x_l[(nx+nu)*(i-1)+5] = min_v
        x_u[(nx+nu)*(i-1)+5] = max_v
        if (i < N)
            x_l[(nx+nu)*(i-1)+6] = min_accel
            x_u[(nx+nu)*(i-1)+6] = max_accel
            x_l[(nx+nu)*(i-1)+7] = min_strrate
            x_u[(nx+nu)*(i-1)+7] = max_strrate          
        end
    end

    # inequality constraint bounds (this is what we do when we have no inequality constraints)
    c_l = zeros(0)
    c_u = zeros(0)
        
    # CAN DONE: inequality constraint bounds
    #c_l = 0.5^2*ones(2*N) # update this 
    #c_u = 1.5^2*ones(2*N) # update this 
    
    # CAN DONE: initialize z0 with the reference Xref, Uref 
    z0 = zeros(idx.nz) # update this
    m = 1
    for x_indx in idx.x
       z0[x_indx] =  Xref2[m]
       m += 1
    end
    p = 1
    for u_indx in idx.u
       z0[u_indx] =  Uref2[p]
       p += 1
    end     
        
    # adding a little noise to the initial guess is a good idea 
    z0 = z0 + (1e-6)*randn(idx.nz)
    
    diff_type = :auto 
    
    Z = fmincon(car_cost,car_equality_constraint,inequality_constraint,
                x_l,x_u,c_l,c_u,z0,params2, diff_type;
                tol = 1e-6, c_tol = 1e-6, max_iters = 10_000, verbose = true)
    
    # pull the X and U solutions out of Z 
    X2 = [Z[idx.x[i]] for i = 1:N2]
    U2 = [Z[idx.u[i]] for i = 1:(N2-1)]
    
    # ------------plotting--------------
    Xm2 = hcat(X2...)
    Um2 = hcat(U2...)
    Xrefm2 = hcat(Xref2...)
    Urefm2 = hcat(Uref2...)
    
    plot(Xrefm2[1,:], Xrefm2[2,:], ls = :dash, label = "reference",
         xlabel = "x (m)", ylabel = "y (m)", title = "Trajectory")  
    display(plot!(Xm2[1,:], Xm2[2,:], label = "actual"))
            
    plot(t_vec2[1:end], Xrefm2[5,:], ls = :dash, label = "reference velocity",
             xlabel = "time (sec)", ylabel = "velocity (m/s)", title = "Velocity Trajectory")  
    display(plot!(t_vec2[1:end], Xm2[5,:], label = "actual"))
        
    plot(t_vec2[1:end], 180/pi*Xrefm2[3,:], ls = :dash, label = "reference heading",
             xlabel = "time (sec)", ylabel = "heading (deg)", title = "Heading Trajectory")  
    display(plot!(t_vec2[1:end], 180/pi*Xm2[3,:], label = "actual"))    

    plot(t_vec2[1:end], 180/pi*Xrefm2[4,:], ls = :dash, label = "reference steering",
             xlabel = "time (sec)", ylabel = "steering (deg)", title = "Steering Trajectory")  
    display(plot!(t_vec2[1:end], 180/pi*Xm2[4,:], label = "actual"))   
    
    plot(t_vec2[1:end-1], Um2[1,:],xlabel = "time (s)", ylabel = "Acceleration [m/s^2]",
                 label = ["Accel"], title = "Acceleration")
    display(plot!(t_vec2[1:end-1], Urefm2[1,:], label = "accel ref"))
        
    plot(t_vec2[1:end-1], 180/pi*Um2[2,:],xlabel = "time (s)", ylabel = "Steering Rate [deg/sec]",
                 label = ["Steer Rate"], title = "Steering Rate")
    display(plot!(t_vec2[1:end-1], Urefm2[2,:], label = "steering rate ref")) 
```

```julia
    # store all our time varying gains in a array of matrices
    K = [zeros(nu,nx) for i = 1:(N+N2)-2]
    # initialize S with Q
    P = [zeros(nx,nx) for i = 1:(N+N2)-1]
    P[N+N2-1] = deepcopy(Qf)
        
    for i = ((N+N2)-2):-1:1
        x = deepcopy(XFULL[:,i])
        u = deepcopy(UFULL[:,i])
        
        A = ForwardDiff.jacobian(_x -> rk4(estimated_car_dynamics, _x, u, dt), x)
        B = ForwardDiff.jacobian(_u -> rk4(estimated_car_dynamics, x, _u,    dt), u)    
        K[i] = (inv(R + transpose(B) * P[i+1] * B) * transpose(B) * P[i+1] * A)
        P[i] = (Q + transpose(A) * P[i+1] * (A - B*K[i]))  
    end # ITERATE THROUGH ALL THE TIME STEPS    
```
